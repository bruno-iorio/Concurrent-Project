```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
csv_file = 'dspd_CAL_results.csv'

df = pd.read_csv(csv_file)

plt.figure(figsize=(10, 6))
for threads in sorted(df['threads'].unique()):
    subset = df[df['threads'] == threads]
    plt.plot(subset['delta'], subset['time_es'], label=f"{threads} threads")

plt.xlabel('bucket size (delta)')
plt.ylabel('runtime in miliseconds')
plt.title('Dynamic Paralell delta stepping - California Roadmap')
plt.legend(title='threads')
plt.tight_layout()
plt.show()
plt.savefig("ds")
```

```python
csv_file1 = 'dspd_RandG_density_results.csv'
csv_file2 = 'dsps_RandG_density_results.csv'
csv_file3 = 'dijkstra_RandG_density_results.csv'

df1 = pd.read_csv(csv_file1)
df2 = pd.read_csv(csv_file2)
df3 = pd.read_csv(csv_file3)

plt.figure(figsize=(10, 6))
plt.plot(df1['density'], df1['time_es'],label='dynamic dp')
plt.plot(df2['density'], df2['time_es'],label='static dp')
plt.plot(df3['density'], df3['time_es'],label='dijkstra')


plt.xlabel('graph density (c)')
plt.ylabel('runtime in miliseconds')
plt.title('Graph density vs. Performance')
plt.legend(title="algorithm")
plt.tight_layout()
plt.show()
```

```python
csv_file1 = 'dspd_CAL_results_fixedthread.csv'
csv_file2 = 'dsps_CAL_results_fixedthread.csv'
csv_file3 = 'dijkstra_CAL_results_fixedthread.csv'

df1 = pd.read_csv(csv_file1)
df2 = pd.read_csv(csv_file2)
df3 = pd.read_csv(csv_file3)

plt.figure(figsize=(10, 6))
plt.bar(0, df1['time_es'],label='dynamic dp')
plt.bar(1, df2['time_es'],label='static dp')
plt.bar(2, df3['time_es'],label='dijkstra')
plt.xticks([])  # Removes both ticks and labels


plt.ylabel('runtime in miliseconds')
plt.title('Parallel delta stepping CAL')
plt.legend(title="algorithm")
plt.tight_layout()
plt.show()
```
