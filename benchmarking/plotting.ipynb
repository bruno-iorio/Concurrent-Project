```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
```

```python
csv_file = 'dsps_RandG_results.csv'
df = pd.read_csv(csv_file)
print(df)
plt.figure(figsize=(10, 6))
for threads in sorted(df['threads'].unique()):
    subset = df[df['threads'] == threads]
    plt.plot(subset['delta'], subset['time_es'], label=f"{threads} threads")

plt.xlabel('bucket size (delta)')
plt.ylabel('runtime in microseconds')
plt.title('Parallel delta stepping (static) on Random Graphs')
plt.legend(title='threads')
plt.tight_layout()
plt.savefig("Parallel delta stepping static.png")
```

```python
df1 = pd.read_csv('dijkstra_RandGdensity_results.csv')
df2 = pd.read_csv('dsps_RandGdensity_results.csv')
df3 = pd.read_csv('dss_RandGdensity_results.csv')
df4 = pd.read_csv('dssd_RandGdensity_results.csv')

x_axis = [k for k in range(2,7)]

y_axis1 = df1['time_es']
y_axis2 = df2['time_es']
y_axis3 = df3['time_es']
y_axis4 = df4['time_es']


plt.figure(figsize=(10, 6))

plt.plot(x_axis,y_axis1,label="dijkstra")
plt.plot(x_axis,y_axis2,label="static parallel delta stepping")
plt.plot(x_axis,y_axis3,label="static sequential delta stepping")
plt.plot(x_axis,y_axis4,label="dynamic sequential delta stepping")
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useOffset=False))
plt.ticklabel_format(style='plain', axis='y')  # optional redundancy
plt.xlabel('density (c)')
plt.ylabel('runtime in microseconds')
plt.title('Graph density vs. performace')
plt.legend(title="algorithm")
plt.tight_layout()
plt.savefig("density_vs_performance.png")
```

```python

df5 = pd.read_csv("dss_RandG_results.csv")
df6 = pd.read_csv("dssd_RandG_results.csv")
print(df5)
plt.figure(figsize=(10, 6))
plt.plot(df5["delta"],df5["time_es"],label="static sequential delta stepping")
plt.plot(df6["delta"],df6["time_es"],label="dynamic sequential delta stepping")
plt.xlabel('bucket size (delta)')
plt.ylabel('runtime in microseconds')
plt.title('Sequential implementation comparison')
plt.legend(title='algorithm')
plt.tight_layout()
plt.savefig("Sequential implementation comparison.png")
```
