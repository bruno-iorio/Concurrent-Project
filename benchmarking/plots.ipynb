```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
csv_file = 'dspd_CAL_results.csv'

df = pd.read_csv(csv_file)

plt.figure(figsize=(10, 6))
for threads in sorted(df['threads'].unique()):
    subset = df[df['threads'] == threads]
    plt.plot(subset['delta'], subset['time_es'], label=f"{threads} threads")

plt.xlabel('bucket size (delta)')
plt.ylabel('runtime in miliseconds')
plt.title('Dynamic Paralell delta stepping - California Roadmap')
plt.legend(title='threads')
plt.tight_layout()
plt.savefig("CAL-dspd.png")
plt.show()

for threads in sorted(df['threads'].unique()):
    subset = df[df['threads'] == threads]
    plt.plot(subset['delta'], 100*subset['cpu_load'], label=f"{threads} threads")

plt.xlabel('bucket size (delta)')
plt.ylabel('CPU Load (%)')
plt.title('CPU Load - Dynamic Parallel Delta Stepping - California Roadmap')
plt.legend(title='threads')
plt.tight_layout()
plt.savefig("CAL-cpu-dspd.png")
plt.show()

for threads in sorted(df['threads'].unique()):
    subset = df[df['threads'] == threads]
    plt.plot(subset['delta'], 100*subset['cpu_load']/threads, label=f"{threads} threads")

plt.xlabel('bucket size (delta)')
plt.ylabel('CPU Load per thread(%)')
plt.title('CPU Load - Dynamic Parallel Delta Stepping - California Roadmap')
plt.legend(title='threads')
plt.tight_layout()
plt.savefig("CAL-cpuperthread-dspd.png")
plt.show()
```

```python
csv_file1 = 'dspd_RandG_density_results.csv'
csv_file2 = 'dsps_RandG_density_results.csv'
csv_file3 = 'dijkstra_RandG_density_results.csv'

df1 = pd.read_csv(csv_file1)
df2 = pd.read_csv(csv_file2)
df3 = pd.read_csv(csv_file3)

plt.figure(figsize=(10, 6))
plt.plot(df1['density'], df1['time_es'],label='dynamic dp')
plt.plot(df2['density'], df2['time_es'],label='static dp')
plt.plot(df3['density'], df3['time_es'],label='dijkstra')


plt.xlabel('graph density (c)')
plt.ylabel('runtime in miliseconds')
plt.title('Graph density vs. Performance')
plt.legend(title="algorithm")
plt.tight_layout()
plt.savefig("dyn Graph-density_vs_performance.png")
plt.show()


plt.figure(figsize=(10, 6))
plt.plot(df2['density'], df2['time_es'],label='static dp')
plt.plot(df3['density'], df3['time_es'],label='dijkstra')


plt.xlabel('graph density (c)')
plt.ylabel('runtime in miliseconds')
plt.title('Graph density vs. Performance')
plt.legend(title="algorithm")
plt.tight_layout()
plt.savefig("without dyn Graph-density_vs_performance.png")
plt.show()

print(df1)
print(df2)
avg_speed_dspd = []
avg_speed_dsps = []
for i in range(0,4):
    avg_speed_dspd.append( df3["time_es"][i+1]/ df1["time_es"][i])
    avg_speed_dsps.append(df3["time_es"][i+1]/ df2["time_es"][i+1])
print("speed up dspd dijsktra: ", avg_speed_dspd)
print("speed up dsps dijsktra: ", avg_speed_dsps)
print("speed up dspd dijsktra: ", sum(avg_speed_dspd)/len(avg_speed_dspd))
print("speed up dsps dijsktra: ", sum(avg_speed_dsps)/len(avg_speed_dsps))
```
